```
import numpy as np
```

```
class Funciones_membresia():
  def __init__(self):
    pass

  def funcion_saturacion(self, x, a, b):
    """
      Funcion Hombro derecho o saturacion

      Funcion que indica el grado de pertenencia de "x" a un conjunto de datos difuso, delimitado por los elementos "a" y "b" donde "a" indica
      el menor grado de pertenencia (0) y "b" el mayor grado de pertenencia (1) al conjunto, "a" debe ser menor que "b" (a < b)

      x = elemento a evaluar
      a = elemeto con menor grado de pertenencia (0), donde empieza recta con inclinacion positiva
      b = elemento con mayor grado de pertenencia (1), donde terminala la recta empezada en el punto "a"
    
    """
    if x <= a:
      result = 0
    elif a < x < b:
      result = (x-a)/(b-a)
    elif x >= b:
      result = 1

    return result

  def funcion_hombro(self, x, a, b):
    """
      Funcion hombro izquierdo

      Funcion que indica el grado de pertenencia de "x" a un conjunto de datos difuso, delimitado por los elementos "a" y "b" donde "a" indica
      el grado de pertenencia mayor (1) y "b" el grado de pertenencia menor (0) al conjunto difuso, "a" debe ser menor que "b" (a < b)

      x = elemento a evaluar
      a = elemeto con mayor grado de pertenencia (1), donde empieza recta con inclinacion negativa
      b = elemento con menor grado de pertenencia (0), donde terminala recta empezada en punto "a"
    
    """
    if x <= a:
      result = 1
    elif a < x < b:
      result = (b-x)/(b-a)
    elif x >= b:
      result = 0

    return result

  def funcion_triangular(self, x, a, m, b):
    """
      Funcion triangular

      Funcion que indica el grado de pertenencia de "x" a un conjunto de datos difuso, delimitado por los elementos "a", "b" y "m", donde "a" y "b" 
      indica el grado de pertenencia menor al cojunto (0) y "m" es el grado de pertenencia mayor al conjunto difuso (1), "a" debe ser menor que "m" y 
      a la vez "m" debe ser menor que "b" (a < m < b)

      x = elemento a evaluar
      a = elemeto con menor grado de pertenencia (0), donde empieza recta con inclinacion positiva 
      m = elemento con mayor grado de pertenencia (1), donde termina recta con empezada en el punto "a" y comienza recta con inclinacion negativa en 
          direccion a al punto "b" 
      b = elemento con menor grado de pertenencia (0), donde terminala recta empezada en el punto "m"
    
    """
    if x <= a or x >= b:
      result = 0
    elif a < x <= m:
      result = (x-a)/(m-a)
    elif m < x < b:
      result = (b-x)/(b-m)

    return result

  def funcion_trapecio(self, x, a, b, c, d):
    """
      Funcion trapecio o pi

      Funcion que indica el grado de pertenencia de "x" a un conjunto de datos difuso, delimitado por los elementos "a", "b", "c" y "d"; donde "a" 
      y "b" indica el grado de pertenencia menor al cojunto difuso (0) y "b" y "c" son el grado de pertenencia mayor al conjunto difuso (1),
      "a < b < c < d".

      x = elemento a evaluar
      a = elemeto con menor grado de pertenencia (0), donde empieza la inclinacion positiva de la recta
      b = elemento con mayor grado de pertenencia (1), donde termina la recta con inclinacion positiva empezada en el punto "a" y comienzo de la 
          recta paralela al eje de los elementos
      c = elemento con mayor grado de pertenencia (1), donde termina la recta paralela al eje de los elementos empezada en el punto "b"
      d = elemento con menor grado de pertenencia (0), donde terminala inclinacion negativa de la recta empezada en el punto "c"
    
    """

    if x <= a or x >= d:
      result = 0
    elif a < x < b:
      result = (x-a) / (b-a)
    elif b <= x <= c:
      result = 1
    elif c < x < d:
      result = (d-x) / (d-c)
    
    return result

  def funcion_sigmoidal(self, x, a, m, b):
    """
      Funcion sigmoidal o s

      Funcion que indica el grado de pertenencia de "x" a un conjunto de datos difuso, delimitado por los elementos "a" y "b", "m" es el punto de 
      inflexi贸n; donde "a" indica el grado de menor pertenencia al cojunto difuso (0), "b" el grado de mayor pertenencia (1), "m" el punto de 
      inflexi贸n calculado normalmento por "m = (a+b)/2", "a" debe ser menor que "m" y al mismo tiempo que "m" debe ser menor que "d" (a < m < d)

      x = elemento a evaluar
      a = elemeto con menor grado de pertenencia (0)
      m = punto de inflexi贸n
      d = elemento con mayor grado de pertenencia (1)
    
    """
    if x <= a:
      result = 0
    elif a < x <= m:
      result = 2 * ((x-a)/(b-a))**2
    elif m < x < b:
      result = 1 - 2 * ((x-b)/(b-a))**2
    elif x >= b:
      result = 1

    return result

def main():
  funciones = Funciones_membresia()
  numeros = np.arange(0, 10, 0.5)
  resultado_saturacion = []
  resultado_hombro = []
  resultado_triangular = []
  resultado_trapecio = []
  resultado_sigmoidal = []

  for i in numeros:
    resultado_saturacion.append(funciones.funcion_saturacion(i,3,5))
    resultado_hombro.append(funciones.funcion_hombro(i,3,5))
    resultado_triangular.append(funciones.funcion_triangular(i,3,5,7))
    resultado_trapecio.append(funciones.funcion_trapecio(i,3,5,7,9))
    resultado_sigmoidal.append(funciones.funcion_sigmoidal(i,3,5,7))
  print ("Conjunto de elementos, desde 0 hasta 9.5 con paso 0.5:\n {}".format(numeros))
  print ("\nEjemplo con funcion de saturaci贸n:\nElementos pertenecientes al conjunto difuso identificado con la funcion de pertenencia\nde saturacion delimitada con los parametros a = 3, b = 5\n: {}".format(resultado_saturacion))
  print ("\nEjemplo con funcion de hombro:\nElementos pertenecientes al conjunto difuso identificado con la funcion de pertenencia\nde hombro delimitada con los parametros a = 3, b = 5\n: {}".format(resultado_hombro))
  print ("\nEjemplo con funcion triangular:\nElementos pertenecientes al conjunto difuso identificado con la funcion de pertenencia\ntriangular, delimitada con los parametros a = 3, m = 5, b = 7\n: {}".format(resultado_triangular))
  print ("\nEjemplo con funcion trapecio:\nElementos pertenecientes al conjunto difuso identificado con la funcion de pertenencia\ntrapecio, delimitada con los parametros a = 3, b = 5, c = 7, d = 9\n: {}".format(resultado_trapecio))
  print ("\nEjemplo con funcion sigmoidal:\nElementos pertenecientes al conjunto difuso identificado con la funcion de pertenencia\nsigmoidal, delimitada con los parametros a = 3, m = 5, b = 7\n: {}".format(resultado_sigmoidal))


if __name__ == '__main__':
	main()
```
